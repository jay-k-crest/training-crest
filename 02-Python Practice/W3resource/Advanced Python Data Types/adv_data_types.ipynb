### None data type

Write a Python function that returns the middle character of a string or "None" if the string length is odd.

```python
def get_middle(str):
    length = len(str)
    if length%2==0:
        return None
    else:
        index = length//2
        return str[index]
print(get_middle('HelloWorld!'))
print(get_middle('HelloWorld'))
```

Named Tuple

Write a Python function that takes a Student named tuple as an argument and calculates the average grade.

```python
from collections import namedtuple
Student = namedtuple("Student",["name","age","marks"])
def calculate_grade(Student):
    return sum(Student.marks)/len(Student.marks)
s1 = Student(name="Jay",age=20,marks=[90,89,68,98,78])
print("average grade :",calculate_grade(s1))
```

Python Counter

Write a python program to find the Jaccard similarity coefficient between two lists using 'Counter' objects.

```python
from collections import Counter
def jacard(list1,list2):
    c1 = Counter(list1)
    c2 = Counter(list2)
    intersection = c1&c2
    union = c1|c2
    
    int_size = sum(intersection.values())
    uni_size = sum(union.values())
    if uni_size == 0:
        return 1.0
    else:
        return int_size/uni_size

a = ['dog', 'cat', 'cat', 'rat']
b = ['dog', 'cat', 'mouse']

similarity = jacard(a,b)
print(similarity)
```
