```python
first_names = [
    "Jay", "Amit", "Ravi", "Neha", "Priya",
    "Ankit", "Kunal", "Sneha", "Rahul", "Pooja"
]

last_names = [
    "Sharma", "Patel", "Mehta", "Verma", "Gupta",
    "Singh", "Joshi", "Iyer", "Khan", "Desai"
]

degrees = [
    "B.Tech", "M.Tech", "B.Sc", "M.Sc",
    "BCA", "MCA", "MBA", "PhD"
]

addresses = [
    "Ahmedabad", "Mumbai", "Delhi", "Pune",
    "Bangalore", "Chennai", "Hyderabad", "Jaipur"
]
```

```python
import random
def random_name_gen():
    fname = random.choice(first_names)
    lname = random.choice(last_names)
    degree = random.choice(degrees)
    address = random.choice(addresses)
    
    return f"{fname} {lname} has degree {degree} and stays at {address}."

print(random_name_gen())
print(random_name_gen())
print(random_name_gen())
print(random_name_gen())
print(random_name_gen())

    
```
