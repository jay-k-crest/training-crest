# Datetime Object Creation

<code>
Write a Pandas program to create:
a) Datetime object for Jan 15 2012. <br>
b) Specific date and time of 9:20 pm. <br>
c) Local date and time.<br>
d) A date without time.<br>
e) Current date.<br>
f) Time from a datetime.<br>
g) Current local time.<br>
</code>


```python
import pandas as pd

a = pd.Timestamp('2012-01-15')
b = pd.Timestamp('2012-01-15 21:20:00')
c = pd.Timestamp.now()
d = pd.Timestamp('2012-01-15').date()
e = pd.Timestamp.today().date()
f = b.time()
g = pd.Timestamp.now().time()

print("a)", a)
print("b)", b)
print("c)", c)
print("d)", d)
print("e)", e)
print("f)", f)
print("g)", g)
```
