# .loc Slicing Based on Row and Column Labels

<code>
Write a Pandas program that uses .loc to slice a DataFrame 
based on row and column labels.
</code>


```python
import pandas as pd
import numpy as np

# Create sample DataFrame
data = {
    'Name': ['Amit', 'John', 'Lisa', 'David', 'Sara'],
    'Math': [85, 90, 78, 88, 92],
    'Science': [89, 94, 80, 84, 95],
    'English': [75, 85, 82, 90, 88]
}

labels = ['r1', 'r2', 'r3', 'r4', 'r5']

df = pd.DataFrame(data, index=labels)

# Slice using .loc (row labels and column labels)
result = df.loc[['r2', 'r4'], ['Name', 'Science']]

print("Original DataFrame:\n")
print(df)

print("\nSliced DataFrame using .loc:\n")
print(result)
```
