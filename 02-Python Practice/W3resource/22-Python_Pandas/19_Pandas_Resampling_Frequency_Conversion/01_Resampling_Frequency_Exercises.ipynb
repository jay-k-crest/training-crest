# Upsample Daily Time Series Data to Hourly Frequency

<code>
Write a Pandas program to upsample daily time series data 
to hourly frequency.
</code>


```python
import pandas as pd
import numpy as np

# Create daily time series data
date_range = pd.date_range(start='2024-01-01', periods=5, freq='D')
data = np.random.randint(10, 100, size=5)

df = pd.DataFrame({'Value': data}, index=date_range)

print("Original Daily Data:\n")
print(df)

# Upsample to hourly frequency
df_hourly = df.resample('H').ffill()

print("\nUpsampled Hourly Data:\n")
print(df_hourly)
```
