# Dictionary to Pandas Series

<code>
Write a Pandas program to convert a dictionary to a Pandas Series.
</code>

Sample Series:

Original dictionary:
{'a': 100, 'b': 200, 'c': 300, 'd': 400, 'e': 800}

Converted series:
a    100  
b    200  
c    300  
d    400  
e    800  
dtype: int64


```python
import pandas as pd

data = {'a': 100, 'b': 200, 'c': 300, 'd': 400, 'e': 800}

series = pd.Series(data)

print(series)
```
