# Create a Subset DataFrame

Create a smaller DataFrame from **movies_metadata.csv** 
by selecting a subset of columns (features).


```python
import pandas as pd

df = pd.read_csv(
    "movies_metadata.csv", 
    encoding='latin1', 
    on_bad_lines='skip', 
    low_memory=False
)


subset_columns = [
    'title', 
    'original_language', 
    'release_date', 
    'budget', 
    'revenue', 
    'runtime', 
    'vote_average', 
    'vote_count', 
    'popularity'
]


subset_df = df[subset_columns].copy()


print(f"Original Shape: {df.shape}")
print(f"Subset Shape: {subset_df.shape}")
print("\nFirst 5 rows of the new Subset DataFrame:")
print(subset_df.head())
```
