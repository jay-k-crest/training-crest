# Horizontal Stacked Bar Plot of Opening and Closing Prices

<code>
Write a Pandas program to create a horizontal stacked bar plot of
opening and closing stock prices of Alphabet Inc. between two specific dates.
</code>


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("alphabet_stock_data.csv", parse_dates=['Date'])

filtered = df[(df['Date'] >= '2020-04-01') & (df['Date'] <= '2020-04-15')]

plot_data = filtered[['Open', 'Close']]
plot_data.index = filtered['Date']

plot_data.plot(kind='barh', stacked=True)

plt.xlabel("Price")
plt.ylabel("Date")
plt.title("Opening and Closing Prices (Alphabet Inc.)")
plt.show()
```
