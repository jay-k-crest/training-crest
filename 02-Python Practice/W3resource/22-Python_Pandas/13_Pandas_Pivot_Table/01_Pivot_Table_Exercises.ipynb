# Manager-Wise Sale Count and Mean Sale Amount

<code>
Write a Pandas program to read the Excel file (Salesdata.xlsx),
create a Pivot Table, and compute the manager-wise sale count
and the mean value of sale amount.
</code>


```python
import pandas as pd

df = pd.read_excel("SaleData.xlsx")

pivot = pd.pivot_table(
    df,
    index='Manager',
    values='Sale_amt',
    aggfunc=['count', 'mean']
)

print(pivot)
```
