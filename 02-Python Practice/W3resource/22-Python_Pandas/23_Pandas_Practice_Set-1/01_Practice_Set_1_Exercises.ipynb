# Summarize Only Object Columns

<code>
Write a Pandas program to summarize only the columns
with data type 'object' from the diamonds DataFrame.
</code>


```python
import pandas as pd

diamonds = pd.read_csv(
    "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv"
)

summary = diamonds.select_dtypes(include='object').describe()

print(summary)
```

# Display Percentages for 'cut' Series Values

Write a Pandas program to display the percentage of each unique value  
in the 'cut' column of the diamonds DataFrame.


```python

cut_percentages = diamonds['cut'].value_counts(normalize=True) * 100

print(cut_percentages)
```
