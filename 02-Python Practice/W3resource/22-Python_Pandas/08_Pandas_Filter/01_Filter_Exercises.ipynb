# Filtering by Year

<code>
Write a Pandas program to find out the alcohol consumption of a given year
from the world alcohol consumption dataset.
</code>

### Test Data

| Year | WHO region       | Country                   | Beverage Types | Display Value |
|------|------------------|--------------------------|----------------|---------------|
| 1986 | Western Pacific  | Viet Nam                 | Wine           | 0.00          |
| 1986 | Americas         | Uruguay                  | Other          | 0.50          |
| 1985 | Africa           | Cte d'Ivoire             | Wine           | 1.62          |
| 1986 | Americas         | Colombia                 | Beer           | 4.27          |
| 1987 | Americas         | Saint Kitts and Nevis    | Beer           | 1.98          |


```python
import pandas as pd

data = {
    'Year': [1986, 1986, 1985, 1986, 1987],
    'WHO region': ['Western Pacific', 'Americas', 'Africa', 'Americas', 'Americas'],
    'Country': ['Viet Nam', 'Uruguay', "Cte d'Ivoire", 'Colombia', 'Saint Kitts and Nevis'],
    'Beverage Types': ['Wine', 'Other', 'Wine', 'Beer', 'Beer'],
    'Display Value': [0.00, 0.50, 1.62, 4.27, 1.98]
}

df = pd.DataFrame(data)

result = df[df['Year'] == 1986]

print(result)
```
