# Grouping by Multiple Columns

<code>
Write a Pandas program to group data by multiple columns 
to perform complex data analysis and aggregations.
</code>


```python
import pandas as pd

# Sample DataFrame
data = {
    'Department': ['IT', 'IT', 'HR', 'HR', 'Finance', 'Finance', 'IT', 'HR'],
    'Employee': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Year': [2022, 2023, 2022, 2023, 2022, 2023, 2022, 2023],
    'Salary': [60000, 65000, 50000, 52000, 70000, 72000, 63000, 54000]
}

df = pd.DataFrame(data)

print("Original Data:\n")
print(df)

# Group by multiple columns
grouped = df.groupby(['Department', 'Year']).agg(
    Total_Salary=('Salary', 'sum'),
    Average_Salary=('Salary', 'mean'),
    Min_Salary=('Salary', 'min'),
    Max_Salary=('Salary', 'max'),
    Employee_Count=('Employee', 'count')
)

print("\nGrouped Data (Department & Year):\n")
print(grouped)
```
