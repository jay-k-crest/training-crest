### Write a Python program to find the latitude and longitude of a given location using Nominatim API and GeoPy package.

```python
from geopy.geocoders import Nominatim  
geolocator =  Nominatim(user_agent='jay')

location = geolocator.geocode("Adajan Surat")
print(location.address)
print(location.latitude, location.longitude)
```

```python
location = geolocator.reverse((21.195299554505606, 72.79261936854617))
print(location.address)
```
