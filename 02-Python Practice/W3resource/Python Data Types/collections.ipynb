###  Find occurrences of the 10 most common words in a text

```python
from collections import Counter
import re

def common_words_with_count(text):
    
    words = re.findall(r"[A-Za-z]+", text)

    counts = Counter(words)

    
    return counts.most_common(10)

text = """
Python is powerful and fast; plays well with others; runs everywhere; is friendly and easy to learn.
Python the Python and We with Software Foundation and PSF. The Python and the Python and the Python.
"""

print(common_words_with_count(text))

```
