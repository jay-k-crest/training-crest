Write a Python program that implements a Python program that filters out dates (in the format "YYYY-MM-DD") that are in the future using the filter function.

```python
from datetime import datetime


dates = [
    "2024-01-01",
    "2026-05-10",
    "2023-12-31",
    "2025-03-15",
    "2022-08-09"
]


today = datetime.today().date()


def is_not_future(date_string):
    date_obj = datetime.strptime(date_string, "%Y-%m-%d").date()
    return date_obj <= today

filtered_dates = list(filter(is_not_future, dates))

print("Original dates:", dates)
print("Dates not in future:", filtered_dates)
```
