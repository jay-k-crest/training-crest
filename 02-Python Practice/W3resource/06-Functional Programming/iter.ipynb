 Generate the running maximum, minimum value of the elements of an iterable

```python
from itertools import accumulate

def running_max(iterable):
    return list(accumulate(iterable, max))
def running_min(iterable):
    return list(accumulate(iterable, min))
```

```python
nums = [5, 2, 7, 1, 9, 3]

print(running_max(nums))  
print(running_min(nums))  
```
