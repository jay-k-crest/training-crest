Write a Python script to display the various Date Time formats.<br>

a) Current date and time <br>
b) Current year<br>
c) Month of year<br>
d) Week number of the year<br>
e) Weekday of the week<br>
f) Day of year<br>
g) Day of the month<br>
h) Day of week<br>

```python
from datetime import datetime

now = datetime.now()

print("Current date and time:", now)
print("Current year:", now.year)
print("Month of year:", now.month)
print("Week number of the year:", now.isocalendar().week)
print("Weekday of the week:", now.isocalendar().weekday)
print("Day of year:", now.timetuple().tm_yday)
print("Day of the month:", now.day)
print("Day of week:", now.strftime("%A"))
```

Write a Python program to print the next 5 days starting today.

```python
import datetime

base = datetime.datetime.today()

for x in range(0,5):
    print(base + datetime.timedelta(days=x))
```
