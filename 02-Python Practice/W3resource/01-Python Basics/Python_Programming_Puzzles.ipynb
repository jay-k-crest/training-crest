# Write a Python program to find the two closest distinct numbers in a given list of numbers. <br>
Input:<br>
<code>[1.3, 5.24, 0.89, 21.0, 5.27, 1.3]</code><br>
Output:<br>
<code>[5.24, 5.27] </code><br>
Input:<br>
<code>[12.02, 20.3, 15.0, 19.0, 11.0, 14.99, 17.0, 17.0, 14.4, 16.8]</code><br>
Output:<br>
<code>[14.99, 15.0]</code><br>

```python
def closest_pair(arr):
    arr.sort()
    min_diff = float('inf')
    a=0
    b=0
    for i in range(1, len(arr)):
        if arr[i] == arr[i-1]:
            continue
        diff = arr[i] - arr[i-1]
        if diff < min_diff:
            min_diff = diff
            a,b = arr[i-1],arr[i]
    return a,b

```

```python
list1 = [1.3, 5.24, 0.89, 21.0, 5.27, 1.3]
list2 = [12.02, 20.3, 15.0, 19.0, 11.0, 14.99, 17.0, 17.0, 14.4, 16.8]
print(closest_pair(list1))
print(closest_pair(list2))
```

# Write a Python program to find the largest negative and smallest positive numbers (or 0 if none).<br>
Input:
[-12, -6, 300, -40, 2, 2, 3, 57, -50, -22, 12, 40, 9, 11, 18]<br>
Output:<br>
[-6, 2]<br>
Input:<br>
[-1, -2, -3, -4]<br>
Output:<br>
[-1, 0]<br>
Input:<br>
[1, 2, 3, 4]<br>
Output:<br>
[0, 1]<br>
Input:<br>
[]<br>
Output:<br>
[0, 0]<br>

```python
def big_small(list):
    if not list:
        return None,None
    negative = [i for i in list if i < 0]
    positive = [i for i in list if i > 0]
    
    biggest_negative = max(negative) if negative else None
    smallest_positive = min(positive) if positive else None

    return biggest_negative, smallest_positive
```

```python
list1= [-12, -6, 300, -40, 2, 2, 3, 57, -50, -22, 12, 40, 9, 11, 18]
list2=[-1, -2, -3, -4]
list3=[1, 2, 3, 4]
list4=[]

print(big_small(list1))
print(big_small(list2))
print(big_small(list3))
print(big_small(list4))
```
