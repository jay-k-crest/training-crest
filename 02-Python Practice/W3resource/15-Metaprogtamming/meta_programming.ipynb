### Python Dynamic Method Addition: Expand Class Behavior

```python
def add_method(cls,func_name,func):
    setattr(cls,func_name,func)

class Myclass:
    pass

def greet(self):
    return "hello from the function which was added through metaprogramming paradigram"
add_method(Myclass,'greet',greet)

obj=  Myclass()
print(obj.greet())
    
    
```
