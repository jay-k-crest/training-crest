NLTK Tokenize: Tokenize a twitter text

```python
from nltk.tokenize import TweetTokenizer

tknizer = TweetTokenizer(strip_handles=True,reduce_len=True)

tweet = "Valar Morghulis : Valar Dohaeris"

tokens = tknizer.tokenize(tweet)
print(tokens)
```

NLTK corpus: Find the sets of synonyms and antonyms of a given word

```python
import nltk
nltk.download('wordnet')
from nltk.corpus import wordnet
```

```python
synonyms = []
antonyms = []

for syn in wordnet.synsets("active"):
    for l in syn.lemmas():
        synonyms.append(l.name()) # Add synonym
        if l.antonyms():
            antonyms.append(l.antonyms()[0].name()) # Add antonym if it exists

print("Synonyms:", set(synonyms))
print("Antonyms:", set(antonyms))
```
