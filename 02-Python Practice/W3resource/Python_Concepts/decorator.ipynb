Create a decorator to convert function return value type

```python
def convert_to_data_type(data_type):
    def decorator(func):
        def wrapper(*args,**kwargs):
            result = func(*args,**kwargs)
            return data_type(result)
        return wrapper
    return decorator

@convert_to_data_type(int)
def add(x,y):
    return x+y

print(add(2,3))
```
