### Recursive Sum Calculation
Calculate the sum of all numbers in a list using a recursive function.

**Logic:**
* **Base Case:** If the list is empty, return 0.
* **Recursive Case:** Add the first number to the result of the function called with the rest of the list.


```python
def get_sum(numbers):
    
    if not numbers:
        return 0
    
    return numbers[0] + get_sum(numbers[1:])

my_list = [1, 2, 3, 4, 5]
print("List:", my_list)
print("Sum:", get_sum(my_list))
```

## üîÅ Recursive Digit Sum

Given a **non-negative integer**, write a Python function that computes the **sum of its digits** using **recursion**.

### üß† Example
- `sumDigits(345)` ‚Üí `12`
- `sumDigits(45)` ‚Üí `9`

The solution must rely purely on recursion without using loops.


```python
def sumDigits(n):
    if n == 0:
        return 0
    return n % 10 + sumDigits(n // 10)

print(sumDigits(345))
print(sumDigits(45))
```
