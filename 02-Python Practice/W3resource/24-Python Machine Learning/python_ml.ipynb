# Iris Dataset - Exercise 1: Visualization

**1. Load the dataset and display:**
   - First 5 rows
   - Dataset shape
   - Basic statistics

**2. Create these plots:**
   - Scatter plot: petal length vs petal width (color by species)
   - Box plot: sepal length for each species
   - Histogram: petal length distribution

**3. Answer:**
   - Which species has the largest petals?
   - Which features separate species best?

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
print("Shape:", df.shape)
print("\nFirst 5 rows:")
print(df.head())
print("\nStats:")
print(df.describe())
```

```python
# 1. Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species')
plt.title('Petal Length vs Width')
plt.show()
```

```python
# 2. Box plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='species', y='sepal_length')
plt.title('Sepal Length by Species')
plt.show()
```

```python
# 3. Histogram
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='petal_length', hue='species', bins=15)
plt.title('Petal Length Distribution')
plt.show()
```

# Iris Dataset - Exercise 2: Train & Test Model

**1. Split data:**
   - Features: petal length & petal width
   - Target: species
   - Test size: 30%, random_state=42

**2. Train KNN model:**
   - Use n_neighbors=3
   - Fit on training data

**3. Evaluate:**
   - Predict on test set
   - Print accuracy score
   - Show confusion matrix

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('iris')
X = df[['petal_length', 'petal_width']]
y = df['species']
```

```python
# Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
# Train
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
# Predict
y_pred = knn.predict(X_test)
```

```python
# Evaluate
print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
