### Write a Python program to compute the sum of the two reversed numbers and display the sum in reversed form. <br>
Input : 13, 14<br>
Output : 27<br>
Note : The result will not be unique for every number for example 31 is a reversed form of several numbers of 13, 130, 1300 etc.<br>
 Therefore all the leading zeros will be omitted<br>

```python
def reversed(num):
    return int(str(num)[::-1])
def func(n1,n2):
    sum = reversed(n1)+reversed(n2)
    result = reversed(sum)
    return result
print(func(13,14))
print(func(1300,21))
```

### Write a Python program to check whether a given number is an ugly number.<br>
Input : 12 <br>
Output : True<br>
Ugly numbers are positive numbers whose only prime factors are 2, 3 or 5. The sequence 1, 2, 3, 4, 5, 6, 8, 9, 10, 12, ...<br>
shows the first 10 ugly numbers.<br>
Note: 1 is typically treated as an ugly number<br>

```python
def ugly(num):
    if num%2 == 0 or num%3==0 or num%5==0:
        return True
    else:
        return False
```

```python
print(ugly(22))
print(ugly(23))  
```
