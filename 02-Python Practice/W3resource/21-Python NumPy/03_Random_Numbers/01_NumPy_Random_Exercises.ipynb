# Minimum and Maximum Values in a Random Array

---

### Problem Statement

<code>Write a NumPy program to create a 5Ã—5 array with random values and find the minimum and maximum values.</code>

---


```python
import numpy as np

arr = np.random.random((5, 5))

print("Original Array:")
print(arr)
print("Minimum and Maximum Values:")
print(arr.min(), arr.max())
```
