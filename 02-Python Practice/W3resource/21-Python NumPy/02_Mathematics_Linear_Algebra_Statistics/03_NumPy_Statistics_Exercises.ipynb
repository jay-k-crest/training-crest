# Cross-Correlation of Two Arrays

---

### Problem Statement

<code>Write a NumPy program to compute cross-correlation of two given arrays.</code>

---

### Sample Output

Original array1:


```python
import numpy as np

arr1 = np.array([0, 1, 3])
arr2 = np.array([2, 4, 5])

result = np.correlate(arr1, arr2, mode='valid')

print("Original array1:")
print(arr1)
print("Original array2:")
print(arr2)
print("Cross-correlation of the said arrays:")
print(result)
```
