# Condition Number of a Matrix

---

### Problem Statement

<code>Write a NumPy program to compute the condition number of a given matrix.</code>

---


```python
import numpy as np

matrix = np.array([[1, 2], [3, 4]])

cond_number = np.linalg.cond(matrix)

print("Matrix:")
print(matrix)
print("Condition Number:")
print(cond_number)
```
