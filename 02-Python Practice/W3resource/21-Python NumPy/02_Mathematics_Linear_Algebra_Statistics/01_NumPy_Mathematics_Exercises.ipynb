# Element-wise Remainder of Division

---

### Problem Statement

<code>Write a NumPy program to get the element-wise remainder of an array of division.</code>

---

### Sample Output

Original array:


```python
import numpy as np

arr = np.arange(7)
result = np.remainder(arr, 5)

print("Original array:")
print(arr)
print("Element-wise remainder of division:")
print(result)
```
