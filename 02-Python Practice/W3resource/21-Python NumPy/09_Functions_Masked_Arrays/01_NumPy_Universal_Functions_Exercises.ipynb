# Row-wise Addition Using NumPy ufunc

---

### Problem Statement

<code>
Write a NumPy program that creates a 2D NumPy array and a 1D array.
Use the np.add ufunc to add the 1D array to each row of the 2D array.
</code>

---


```python
import numpy as np

arr2d = np.array([[1, 2, 3],
                  [4, 5, 6]])

arr1d = np.array([10, 20, 30])

result = np.add(arr2d, arr1d)

print(result)
```
