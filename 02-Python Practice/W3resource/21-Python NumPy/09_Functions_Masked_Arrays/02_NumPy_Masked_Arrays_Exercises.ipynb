# Creating a Masked Array Based on Condition

---

### Problem Statement

<code>
Write a NumPy program that creates a masked array where all elements greater than a specified value are masked.
</code>

---


```python
import numpy as np

arr = np.array([10, 25, 30, 15, 50, 5])

masked = np.ma.masked_greater(arr, 20)

print(masked)
```
