# Saving and Loading NumPy Arrays from Text Files

---

### Problem Statement

<code>
Write a NumPy program that saves a NumPy array to a text file with space as the delimiter and then reads it back into a NumPy array.
</code>

---


```python
import numpy as np

arr = np.array([[1, 2, 3],
                [4, 5, 6]])

np.savetxt("array.txt", arr, delimiter=" ")

loaded = np.loadtxt("array.txt")

print(loaded)
```
