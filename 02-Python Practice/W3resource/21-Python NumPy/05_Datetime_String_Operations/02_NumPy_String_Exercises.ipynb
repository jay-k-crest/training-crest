# Encoding and Decoding Using cp500

---

### Problem Statement

<code>
Write a NumPy program to encode all the elements of a given array in cp500 and decode it again.
</code>

---

### Sample Output

encoded =


```python
import numpy as np

arr = np.array(['python exercises', 'PHP', 'java', 'C++'])

encoded = np.char.encode(arr, 'cp500')
decoded = np.char.decode(encoded, 'cp500')

print("encoded =", encoded)
print("decoded =", decoded)
```
