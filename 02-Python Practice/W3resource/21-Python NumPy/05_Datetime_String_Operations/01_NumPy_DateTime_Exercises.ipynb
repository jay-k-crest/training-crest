# First Monday in May 2017

---

### Problem Statement

<code>Write a NumPy program to find the first Monday in May 2017.</code>

---

### Expected Output

First Monday in May 2017:


```python
import numpy as np

dates = np.arange('2017-05-01', '2017-06-01', dtype='datetime64[D]')
first_monday = dates[np.where(dates.astype('datetime64[W]') == dates.astype('datetime64[D]'))][0]

print("First Monday in May 2017:")
print(first_monday)
```
