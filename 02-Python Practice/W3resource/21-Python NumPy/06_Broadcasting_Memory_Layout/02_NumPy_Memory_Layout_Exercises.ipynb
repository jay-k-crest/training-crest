# 3D Array Strides and Column-Major Flattening

---

### Problem Statement

<code>
Write a NumPy program that creates a 3D array of shape (2, 4, 4) and print the strides.
Then, flatten it with ravel(order='F') and print the flattened array.
</code>

---


```python
import numpy as np

arr = np.arange(32).reshape(2, 4, 4)

print("Strides:")
print(arr.strides)

flattened = arr.ravel(order='F')

print("Flattened array (Fortran order):")
print(flattened)
```
