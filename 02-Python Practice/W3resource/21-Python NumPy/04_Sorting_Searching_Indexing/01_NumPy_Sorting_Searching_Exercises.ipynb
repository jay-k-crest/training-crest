# Sorting Students by Height

---

### Problem Statement

<code>
Write a NumPy program to sort the student id with increasing height of the students from given students id and height.
Print the integer indices that describes the sort order by multiple columns and the sorted data.
</code>

---


```python
import numpy as np

student_id = np.array([1023, 5202, 6230, 1671, 1682, 5241, 4532])
height = np.array([40.0, 42.0, 45.0, 41.0, 38.0, 40.0, 42.0])

indices = np.lexsort((student_id, height))
sorted_data = np.column_stack((student_id[indices], height[indices]))

print("Sorted indices:")
print(indices)
print("Sorted data:")
for row in sorted_data:
    print(int(row[0]), row[1])
```
