# Sorting a Structured Array by Height

---

### Problem Statement

<code>
Write a NumPy program that sorts the structured array created with fields for 'name' (string), 
'age' (integer), and 'height' (float), by the 'height' field in ascending order and prints the sorted array.
</code>

---


```python
import numpy as np

dtype = [('name', 'U10'), ('age', 'i4'), ('height', 'f4')]

data = np.array([
    ('Alice', 25, 5.5),
    ('Bob', 22, 5.9),
    ('Charlie', 23, 5.7),
    ('David', 24, 5.6)
], dtype=dtype)

sorted_data = np.sort(data, order='height')

print(sorted_data)
```
