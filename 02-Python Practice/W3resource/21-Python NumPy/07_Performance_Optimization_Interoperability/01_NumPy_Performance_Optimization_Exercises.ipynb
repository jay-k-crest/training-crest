# Finding Maximum Element: Loop vs NumPy

---

### Problem Statement

<code>
Write a NumPy program to generate a large NumPy array and write a function to find the maximum element using a for loop.
Optimize it using NumPy's built-in functions.
</code>

---


```python
import numpy as np

arr = np.random.random(1000000)

def max_with_loop(a):
    m = a[0]
    for v in a:
        if v > m:
            m = v
    return m

print("Maximum using loop:")
print(max_with_loop(arr))

print("Maximum using NumPy:")
print(arr.max())
```
