#  NumPy Array & Memory Footprint

---

###  Problem Statement

<code>Write a NumPy program to create an array with the values 1, 7, 13, 105 and determine the size of the memory occupied by the array.</code>

---


```python
import numpy as np

arr = np.array([1, 7, 13, 105])

print(arr)
print(arr.nbytes)
```
