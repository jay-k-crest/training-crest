Write a Python program using the math module to:

Find the square root of a number

Calculate the factorial of another number

Find the value of sin(90Â°)

```python
import math


num1 = 16
num2 = 5


sqrt_value = math.sqrt(num1)


factorial_value = math.factorial(num2)

sin_90 = math.sin(math.radians(90))


print("Square root of", num1, ":", sqrt_value)
print("Factorial of", num2, ":", factorial_value)
print("sin(90 degrees):", sin_90)
```
