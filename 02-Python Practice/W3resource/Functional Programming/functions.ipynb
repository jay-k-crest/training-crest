Calculate the number of upper / lower case letters in a string

```python
def str_test(s):
    uppercase = 0
    lowercase = 0
    
    for c in s:
        if c.isupper():
            uppercase+=1
        elif c.islower():
            lowercase+=1
        else:
            pass
    return [uppercase,lowercase]
```

```python
s = "This is a test String For the CApital Small Function test"

result = str_test(s)
print(f"The Uppercase letters are : {result[0]} and lowercase are  : {result[1]}")
```
