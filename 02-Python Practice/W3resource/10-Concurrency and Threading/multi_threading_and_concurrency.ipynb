Write a Python program to calculate the factorial of a number using multiple threads.

```python
import threading

def factorial(n):
    result = 1
    for i in range(1,n+1):
        result *=i
    return result

def fact(n):
    print(f"current thread for {n} : {threading.current_thread().name}")
    r = factorial(n)
    print(f"factorial is : {r}")
n=12
t1=threading.Thread(target=fact,args=(n,))
t2=threading.Thread(target=fact,args=(n,))

t1.start()
t2.start()

t1.join()
t2.join()
```

Write a Python program that performs concurrent HTTP requests using threads.

```python
import requests
import threading
def make_request(url):
    response = requests.get(url)
    print(f"Response from {url}: {response.status_code}")
# List of URLs to make requests to
urls = [
    "https://www.example.com",
    "https://www.google.com",
    "https://www.wikipedia.org",
    "https://www.python.org"
]
# Create and start threads for each URL
threads = []
for url in urls:
    thread = threading.Thread(target=make_request, args=(url,))
    thread.start()
    threads.append(thread)
# Wait for all threads to finish
for thread in threads:
    thread.join()
```
