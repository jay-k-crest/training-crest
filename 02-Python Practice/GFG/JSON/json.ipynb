# Read JSON from a File
Write a Python program to read data from a JSON file named <code>student.json</code> using the <code>json</code> module.<br>
Load the JSON content into a Python dictionary and print the dictionary on the screen.

```python
import json

with open("./files/student.json", "r") as file:
    data = json.load(file)

print(data)
```

# Convert Python Dictionary to JSON String
Create a Python dictionary containing student details like <code>name</code>, <code>age</code>, and <code>marks</code>.<br>
Write a program to convert this dictionary into a JSON formatted string using <code>json.dumps()</code> and print it.

```python
import json

student = {
    "name": "jay",
    "age": 20,
    "marks": 88
}

json_string = json.dumps(student)

print(json_string)
```

# Write Dictionary Data to JSON File
Create a Python dictionary representing product details such as <code>id</code>, <code>name</code>, and <code>price</code>.<br>
Write a program to save this dictionary into a JSON file named <code>product.json</code> using <code>json.dump()</code>.

```python
import json

product = {
    "id": 101,
    "name": "Laptop",
    "price": 55000
}

with open("./files/product.json", "w") as file:
    json.dump(product, file)

```

# Parse JSON String and Access Values
Given a JSON string containing student information, convert it to a Python dictionary using <code>json.loads()</code>.<br>
Access and print individual fields such as <code>name</code> and <code>age</code> from the converted dictionary.

```python
import json

json_data = '{"name": "jay", "age": 20, "marks": 92}'

student = json.loads(json_data)

print("Student Name:", student["name"])
print("Student Age:", student["age"])
```

# Update a Value Inside JSON Data
Read data from <code>student.json</code> and load it as a Python dictionary.<br>
Modify one of the values, such as changing the studentâ€™s <code>marks</code> or <code>age</code>.<br>
Write the updated dictionary back into the same JSON file.

```python
import json


with open("./files/student.json", "r") as file:
    students = json.load(file)


for s in students:
    if s["id"] == 1:
        s["age"] = 22


with open("./files/student.json", "w") as file:
    json.dump(students, file, indent=4)

print("after")
with open("./files/student.json", "r") as file:
    data = json.load(file)

print(data)
```

# Work with List of JSON Objects
Assume a JSON file contains a list of employee objects, each having <code>name</code> and <code>salary</code>.<br>
Write a Python program to read this list, find employees having salary greater than a given amount, and print their names.

```python
import json

limit = 50000

with open("./files/employees.json", "r") as file:
    employees = json.load(file)

print("Employees with salary greater than", limit, ":")
for emp in employees:
    if emp["salary"] > limit:
        print(emp["name"])
```

# Pretty Print JSON Data
Write a Python program that reads JSON data from a file and prints it in a properly formatted way (indentation).<br>
Use <code>json.dumps()</code> with the <code>indent</code> parameter to display the output neatly.

```python
import json

with open("./files/student.json", "r") as file:
    data = json.load(file)

pretty_json = json.dumps(data, indent=4)


print(pretty_json)
```
