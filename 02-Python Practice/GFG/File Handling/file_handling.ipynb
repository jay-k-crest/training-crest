# Create and Write to a Text File
Write a Python program to create a text file named <code>notes.txt</code> and write a few lines of text into it using file handling.<br>
After writing, open the same file again and display its contents on the screen

```python

file_path = "./files/notes.txt"


with open(file_path, "w") as f:
    f.write("Hello! These are some notes.\n")
    f.write("We are learning Python file handling.\n")
    f.write("This text is being written into notes.txt.\n")

with open(file_path, "r") as f:
    content = f.read()
print(content)
```

# Read a File Line by Line
Write a Python program to read a text file line by line using a <code>for</code> loop.<br>
Store each line in a list named <code>lines</code> after removing the newline character.<br>
Print the list of lines.

```python


file_path = "./files/notes.txt"

lines = []


with open(file_path, "r") as f:
    for line in f:
        lines.append(line.strip())  

print("Lines stored inside the list:")
print(lines)
```

# Count Words in a Text File
Write a Python program to count the total number of words present in a file named <code>story.txt</code>.<br>
Open the file, read its contents, split the text into words, and print the word count.

```python


file_path = "./files/story.txt"

with open(file_path, "r") as f:
    text = f.read()

words = text.split()

print("Total number of words:", len(words))
```

# Append Content to an Existing File
Create a file named <code>log.txt</code> if it does not already exist.<br>
Write a Python program to append new text to the same file without overwriting existing content.<br>
Display the full file content after appending.

```python


file_path = "./files/log.txt"

with open(file_path, "a") as f:
    f.write("Another log entry has been added.\n")

with open(file_path, "r") as f:
    content = f.read()

print("Current content of log.txt:")
print(content)
```

# Copy Content from One File to Another
Write a Python program to read data from <code>source.txt</code> and copy it into <code>destination.txt</code>.<br>
Ensure that the contents of the destination file become exactly the same as the source file.

```python


source = "./files/source.txt"
destination = "./files/destination.txt"
with open(source, "r") as src:
    data = src.read()

with open(destination, "w") as dest:
    dest.write(data)

print("File copied successfully")
```

# Count Number of Lines, Words, and Characters
Write a Python program to open a text file and count:<br>

Total number of lines<br>

Total number of words<br>

Total number of characters (including spaces)<br>
Print all three counts separately.

```python


file_path = "./files/source.txt"

with open(file_path, "r") as f:
    text = f.read()

total_lines = len(text.splitlines())
total_words = len(text.split())
total_characters = len(text)


print("Lines      :", total_lines)
print("Words      :", total_words)
print("Characters :", total_characters)
```

# Remove Blank Lines from a File
Write a Python program to read a text file and create a new file where all blank lines are removed.<br>
The original file should remain unchanged; only the new cleaned file should be saved.

```python


source = "./files/story.txt"
cleaned = "./files/story_cleaned.txt"

with open(source, "r") as src, open(cleaned, "w") as out:
    for line in src:
        # write only non-empty lines
        if line.strip():
            out.write(line)


```
