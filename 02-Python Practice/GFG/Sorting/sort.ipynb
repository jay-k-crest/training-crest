# Bubble Sort
Write a Python function <code>bubble_sort(arr)</code> to sort a list of integers in ascending order using the Bubble Sort algorithm.<br>
The function should repeatedly swap adjacent elements if they are in the wrong order.<br>
Print the list before and after sorting.

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

arr = [64, 34, 25, 12, 22, 11, 90]
print(bubble_sort(arr))
```

# Insertion Sort
Write a Python function <code>insertion_sort(arr)</code> to sort a list of integers using the Insertion Sort algorithm.<br>
Insert each element into its correct position in the already sorted part of the list.<br>
Print the list before and after applying insertion sort.

```python
def insertion_sort(arr): 
    for i in range(1, len(arr)): 
        key = arr[i] 
        j = i-1
        while j >=0 and key < arr[j] : 
                arr[j+1] = arr[j] 
                j -= 1
        arr[j+1] = key
    return arr
arr = [12, 11, 13, 5, 6]
print("Sorted array is:")
print(insertion_sort(arr))
```

# Descending Order Sort
Modify any sorting algorithm to sort a list of numbers in descending order.<br>
Write a function <code>sort_desc(arr)</code> that returns the sorted list in descending order.<br>
Do not use built-in <code>sort()</code> or <code>sorted()</code> functions.

```python
def sort_desc_bubble(arr):
    for i in range(len(arr)):
        for j in range(0, len(arr) - i - 1):
            if arr[j] < arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

arr = [64, 34, 25, 12, 22, 11, 90]
print("Sorted array is: ", sort_desc_bubble(arr))
    
```

# Merge Sort (Recursive)
Write a Python function <code>merge_sort(arr)</code> to sort a list of integers using the Merge Sort algorithm.<br>
Recursively divide the list into two halves and then merge them in sorted order.<br>
Print the final sorted list.

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left = arr[:mid]
        right = arr[mid:]

        merge_sort(left)
        merge_sort(right)

        i = j = k = 0

        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                arr[k] = left[i]
                i += 1
            else:
                arr[k] = right[j]
                j += 1
            k += 1

        while i < len(left):
            arr[k] = left[i]
            i += 1
            k += 1

        while j < len(right):
            arr[k] = right[j]
            j += 1
            k += 1

    return arr

arr = [12, 11, 13, 5, 6, 7]
print(merge_sort(arr))
```
