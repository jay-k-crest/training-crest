⚙️ Lambda, Map, Filter, and Scope in Python

These concepts help you write code that is short, powerful, and expressive:

Lambda → small, one-line anonymous functions

Map → applies a function to every item in a sequence

Filter → keeps only items that meet a condition

Scope → decides where variables live and can be accessed

```python
def squared(x):
    return x**2
```

```python
my_nums = [1,2,3,4,5]
```

```python
new_list = list(map(squared, my_nums))
new_list
```

```python
def splicer(mystring):
    if len(mystring) % 2 == 0:
        return 'EVEN'
    else:
        return mystring[0]
```

```python
names = ['jay','harsh','aayushi','kavya','john','sandy']
```

```python
list(map(splicer, names))
```

```python
def check_even(num):
    return num % 2 == 0
```

```python
mynums = [1,2,3,4,5,6]
```

```python
list(filter(check_even, mynums))
```

```python
def squared(x):
    return x**2
```

```python
squared(3)
```

```python
squared = lambda x: x**2
```

```python
squared(3)
```

```python
list(map(lambda x: x**2, my_nums))
```

```python
list(filter(lambda num: num % 2 == 0, mynums))
```

```python
names
```

```python
list(map(lambda x:x[0] if len(x)%2 !=0 else 'EVEN', names))
```

LEGB Rule in Python

The LEGB rule explains how Python finds variable values.
It searches in this order:

L → Local → inside the current function
E → Enclosing → inside any outer functions
G → Global → at the top level of the script/module
B → Built-in → Python’s built-in names

```python
# lambda x: x**2 local variable 
```

```python
name = 'this is a global name'
#global variable
def greet():
    #enclosing variable
    name = 'jay'
    def hello():
        #local variable
        print('Hello ' + name)
    hello()
```

```python
greet()
```

```python
x=50
def func():
    global x
    print(f'X is {x}')

    #local reassignment
    x = 200
    print(f'I just locally changed X to {x}')
```

```python
func(x)
```

```python
print(x)
```

```python
func()
```
