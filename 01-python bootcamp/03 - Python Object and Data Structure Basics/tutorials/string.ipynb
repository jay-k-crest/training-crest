```python
mystr = 'abcdefghi'
```

```python
mystr[2:]
#[n:] start from n to end
```

String formatting for printing 

2 methods are there 
1 is .format()  & 
2nd is f-string method 


```python
print('This is a string {}'.format('inserted'))
```

```python
print('The {} {} {}'.format('fox','brown','fast'))
```

```python
print('The {0} {2} {1}'.format('fox','brown','fast'))
```

```python
print('The {f} {q} {b}'.format(f='fox',b='brown',q='fast'))
#indexing also works'
```

Float formatting

```python
r = 100/777
```

```python
r
```

```python
print('the result is {r}'.format(r=r))
```

{VALUE:WIDTH.PRECISION F}

```python
print('the result is {r:1.3f}'.format(r=r))
```

```python
result  = 104.12345
```

```python
print('the result is {r:1.2f}'.format(r=result))
```
