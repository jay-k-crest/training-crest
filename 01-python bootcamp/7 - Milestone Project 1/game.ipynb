```python
board = [
    [' ', ' ', ' '],
    [' ', ' ', ' '],
    [' ', ' ', ' ']
]
```

```python
def print_board(board):
    num = 1
    print()
    for r in range(3):
        row_out = []
        for c in range(3):
            if board[r][c] == ' ':
                row_out.append(str(num))
            else:
                row_out.append(board[r][c])
            num += 1

        print(" {} | {} | {} ".format(*row_out))
        if r < 2:
            print("---+---+---")
    print()

```

```python
print_board(board)
```

```python
pos_map = {
    1:(0,0), 2:(0,1), 3:(0,2),
    4:(1,0), 5:(1,1), 6:(1,2),
    7:(2,0), 8:(2,1), 9:(2,2)
}
```

```python
def place_mark(board, position, mark):
    r, c = pos_map[position]
    if board[r][c] == ' ':
        board[r][c] = mark
        return True
    else:
        return False   # spot already taken
```

```python
def check_winner(board):
    # rows
    for r in range(3):
        if board[r][0] == board[r][1] == board[r][2] != ' ':
            return board[r][0]

    # columns
    for c in range(3):
        if board[0][c] == board[1][c] == board[2][c] != ' ':
            return board[0][c]

    # diagonals
    if board[0][0] == board[1][1] == board[2][2] != ' ':
        return board[0][0]

    if board[0][2] == board[1][1] == board[2][0] != ' ':
        return board[0][2]

    return None
```

```python
def is_draw(board):
    for row in board:
        if ' ' in row:
            return False
    return True
```

```python
current_player = 'X'  # X always starts

while True:
    
    print_board(board)
    print("Player", current_player)
    
    # take input
    try:
        pos = int(input("Choose position (1-9): "))
    except:
        print("Enter a number!")
        continue
    
    # valid range check
    if pos not in range(1,10):
        print("Invalid position. Try again.")
        continue
    
    # place mark
    if not place_mark(board, pos, current_player):
        print("Position already taken. Try again.")
        continue

    # check winner
    winner = check_winner(board)
    if winner:
        print_board(board)
        print("Player", winner, "wins!")
        break

    # check draw
    if is_draw(board):
        print_board(board)
        print("Game is a draw!")
        break

    # switch player
    current_player = 'O' if current_player == 'X' else 'X'
7
```
