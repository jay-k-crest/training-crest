Class Object attributes

```python
class Dog:
    #class object attribute
    species = "Canis lupus familiaris"
    def __init__(self, breed,name,age):
        #self keyword similar to this in java or C++
        self.breed = breed
        self.name = name
        self.age = age
    def bark(self, sound):
        return f"{self.name} says {sound}"
```

```python
my_dog = Dog(breed='Labrador',name='Sammy',age=3)
```

```python
my_dog.bark("Woof Woof")
```

```python
class Circle:
    pi = 3.14

    def __init__(self, radius=1):
        self.radius = radius

    def area(self):
        return self.pi * (self.radius ** 2)

    def circumference(self):
        return 2 * self.pi * self.radius
```

```python
my_circle = Circle()
```

```python
my_circle.pi
```

```python
my_circle.radius
```

```python
my_circle = Circle(30)
```

```python
my_circle.circumference()
```
