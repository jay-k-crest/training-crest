```python

text = input("Enter text: ")
key = int(input("Enter key: "))
```

```python
encrypted = ""

for ch in text:
    if ch.islower():
        pos = ord(ch) - ord('a')
        new_pos = (pos + key) % 26
        encrypted += chr(new_pos + ord('a'))

    elif ch.isupper():
        pos = ord(ch) - ord('A')
        new_pos = (pos + key) % 26
        encrypted += chr(new_pos + ord('A'))

    else:
        encrypted += ch

print("Encrypted text:", encrypted)
```

```python
decrypted = ""

for ch in encrypted:
    if ch.islower():
        pos = ord(ch) - ord('a')
        new_pos = (pos - key) % 26
        decrypted += chr(new_pos + ord('a'))

    elif ch.isupper():
        pos = ord(ch) - ord('A')
        new_pos = (pos - key) % 26
        decrypted += chr(new_pos + ord('A'))

    else:
        decrypted += ch

print("Decrypted text:", decrypted)
```
