Timeit Module

The timeit module measures how long small pieces of Python code take to run. It is useful for comparing different approaches and checking which solution is faster.


```python
def func1(n):
    return [str(num) for num in range(n)]
```

```python
def func2(n):
    return list(map(str,range(n)))
```

```python
func1(10)
```

```python
func2(10)
```

both give same output so messure time

```python
#current time before
import time
start_time = time.time()
#run code
result = func1(100000)
#after time
end_time = time.time()
#differece
total_time = end_time-start_time
print(total_time)
```

```python
#current time before
import time
start_time = time.time()
#run code
result = func2(100000)
#after time
end_time = time.time()
#differece
total_time = end_time-start_time
print(total_time)
```

```python
import timeit
```

```python
stmt = '''
func1(1000)
'''
```

```python
setup = ''' 
def func1(n):
    return [str(num) for num in range(n)]
'''
```

```python
timeit.timeit(stmt,setup,number=10000)
```
