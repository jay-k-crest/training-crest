Zipping and Unzipping Modules

Zipping combines multiple sequences into pairs or groups, while unzipping separates them again. This is often done with the zip() function to bundle and unbundle related data.

```python
with open('../files/zip.txt','w+') as f:
    f.write('one file')
```

```python
with open('../files/zip2.txt','w+') as f:
    f.write('one file')
```

```python
import zipfile
```

```python
comp_file = zipfile.ZipFile('comp_file.zip','w')
```

```python
comp_file.write('../files/zip.txt',compress_type=zipfile.ZIP_DEFLATED)
```

```python
comp_file.write('../files/zip2.txt',compress_type=zipfile.ZIP_DEFLATED)
```

```python
comp_file.close()
```

```python
zip_obj = zipfile.ZipFile('comp_file.zip','r')
```

```python
zip_obj.extractall('extracted_content')
```

```python
import shutil
```

```python
dir_to_zip = './extracted_content/'
```

```python
output_file = 'example'
```

```python
shutil.make_archive(output_file,'zip',dir_to_zip)
```

```python
shutil.unpack_archive('example.zip','final_unzip','zip')
```
