## Task One: Grab the Google Drive Link from .csv File

```python
import csv
```

```python
data = open('../files/find_the_link.csv',encoding='utf-8')
```

```python
csv_data = csv.reader(data)
```

```python
data_lines = list(csv_data)
```

```python
link = []
for n in range(len(data_lines)):
    link.append(data_lines[n][n])
final_link = "".join(link)
final_link
```

## Task Two: Download the PDF from the Google Drive link and find the phone number that is in the document. 

```python
import PyPDF2
import re
```

```python
f = open('../files/Find_the_Phone_Number.pdf', 'rb')
```

```python
pdf_reader = PyPDF2.PdfReader(f)
```

```python
len(pdf_reader.pages)
#page_one = pdf_reader.pages[0]
```

```python
# You should get this phone number
# 505 503 4455
pattern = r'\d{3}.\d{3}.\d{4}'
for n in range(len(pdf_reader.pages)):
    page = pdf_reader.pages[n]
    text = page.extract_text()
    match = re.search(pattern,text)
     
    if match:
        print(match.group())
```
