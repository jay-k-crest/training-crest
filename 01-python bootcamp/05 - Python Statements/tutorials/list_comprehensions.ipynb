List comprehensions provide a short, elegant way to create lists.
They let you build a new list from an existing one in a single readable line.

Use them when you want your code to be:

more compact

more Pythonic

easier to read and modify

Think of list comprehensions as:
üëâ ‚ÄúLoop + condition + result ‚Äî all in one line.‚Äù

```python
mystring = "Hello World"
```

```python
mylist = []
for letter in mystring:
    mylist.append(letter)
mylist
```

```python

mylist = [letter for letter in mystring]
mylist
```

```python
list2 = [x for x in 'Hello']
list2
```

```python
list3 = [num for num in range(0,11)]
list3
```

```python
list3 = [num**2 for num in range(0,11)]
list3
```

```python
list4 = [num for num in range(0,11) if num%2==0]
list4
```

```python
celcius = [0,10,20,34.5]
fahrenheit = [((9/5)*temp + 32) for temp in celcius]
fahrenheit
```

```python
results = [x if x%2==0 else 'ODD' for x in range(0,11)]
results
```

```python
mylist = []
for x in [2,4,6]:
    for y in [100,200,300]:
        mylist.append(x*y)
mylist
```

```python
mylist = [x*y for x in [2,4,6] for y in [100,200,300]]
mylist
```
