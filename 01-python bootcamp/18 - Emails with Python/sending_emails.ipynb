## Sending Emails in Python (SMTP) â€” Keywords Table

| Keyword / Method | Meaning |
|------------------|--------|
| `smtplib`        | Library used to send emails |
| `SMTP()`         | Connect to SMTP server |
| `ehlo()`         | Identify yourself to the server |
| `starttls()`     | Enable encryption (TLS) |
| `login()`        | Login using email + app password |
| `sendmail()`     | Send the email |
| `quit()`         | Close the SMTP connection |


```python
import smtplib
import getpass
```

```python
ob = smtplib.SMTP('smtp.gmail.com',587)
```

```python
ob.ehlo() 
```

```python
ob.starttls()
```

```python
email = getpass.getpass("Enter your email: ")
password = getpass.getpass("Enter your password: ")
ob.login(email,password)
```

```python
from_adderss = email
to_adderess = email
subject = input("enter subject : ")
body = input("enter body : ")
msg = "Subject: "+subject+'\n'+body
ob.sendmail(from_adderss,to_adderess,msg)
```

```python
ob.quit()
```
