Try, Except, and Finally in Python

try, except, and finally are used for error handling in Python. They allow code to continue running even when something goes wrong.

try — the block where you place code that may cause an error

except — runs only if an error occurs, used to handle the error

finally — always runs at the end, whether an error happened or not

These blocks help write programs that are safer, more controlled, and do not stop suddenly due to unexpected errors.

```python
def add(a,b):
    return a + b
```

```python
add(2,3)
```

```python
number1 = 10
```

```python
number2 = input("Please enter a number: ")
```

```python
add(number1, number2)
#error coz number 2 is string
print("after   the error block")
#print will not run
```

```python
try:
    result = add(number1, number2)
    print("The result is:", result)
except TypeError as e:
    print("A TypeError occurred:", e)
print("Continuing with the rest of the program.")
```

```python
#try except finally
a=5
b=10
try:
    result = add(a, b)
    print("The result is:", result)
except TypeError as e:
    print("A TypeError occurred:", e)
finally:
    print("This will always execute regardless of errors.")
```

```python
#try except finally
a=5
b='10'
try:
    result = add(a, b)
    print("The result is:", result)
except TypeError as e:
    print("A TypeError occurred:", e)
finally:
    print("This will always execute regardless of errors.")
```

```python
def ask_for_int():
    while True:
        try:
            result = int(input("Please enter an integer: "))
            print("You entered:", result)
        except ValueError:
            print("That's not a valid integer!")
            continue
        else:
            print("Thank you for entering a valid integer.")
            break
        finally:
            print("This will always execute regardless of errors.")
```
