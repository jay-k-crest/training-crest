# check for 2 star and title grabing done
# ex.select('a')[1]['title']

```python
two_star_titles = []

for n in range(1,51):
    s_url = link.format(n)
    res = requests.get(s_url)
    soup = bs4.BeautifulSoup(res.text,'lxml')
    books = soup.select('.product_pod')
    
    for b in books:
        if not b.select('.star-rating.Two'):
           continue  
        else:
            two_star_titles.append(b.select('a')[1]['title'])   
           

```
