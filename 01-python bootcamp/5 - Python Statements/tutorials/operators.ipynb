range() → creates a sequence of numbers (great for loops)

enumerate() → gives both index and value while looping

zip() → pairs items from multiple sequences together

in → checks if a value is present in a sequence

min() / max() → find the smallest and largest values

random library → used to generate random numbers & choices

input() → takes user input from the keyboard

```python
index_counter = 0
word = 'abcde'
for letter in word:
    print(f'At index {index_counter} the letter is {letter}')
    index_counter += 1
```

```python

word = 'abcde'
for index,letter in enumerate(word):
    print(index)
    print(letter)
    print('\n')
```
