If, Elif, and Else Statements in Python

Conditional statements are used to make decisions in a program.
They allow Python to execute different blocks of code based on whether a condition is true or false.

if checks the first condition,
elif checks additional conditions,
else runs when none of the previous conditions are true.

```python
if True:
    print("This is true")
```

```python
if 3>2:
    print("This is true")
```

```python
hungry = False
if hungry:
    print("Feed me!")   
else:
    print("I'm not hungry")
```

```python
loc  = 'Bank'
if loc == 'Auto Shop':
    print("Cars are cool!")
elif loc == 'Bank':
    print("Money is cool!")
else:
    print("Welcome to the store")
```

```python
name = 'Jay'
if name == 'Frankie':
    print("Hello Frankie")  
elif name == 'Jay':
    print("Hello Jay")            
else:
    print("What is your name?")
```
