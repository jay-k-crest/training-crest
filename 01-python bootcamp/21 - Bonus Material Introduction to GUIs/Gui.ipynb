```python
from ipywidgets import interact,interactive,fixed
import ipywidgets as widgets
```

```python
def func(x):
    return x
```

```python
interact(func,x=10)
```

```python
interact(func,x=False)
```

```python
interact(func,x="hello")
```

```python
@interact(x=True,y=fixed(1.0))
def g(x,y):
    return (x,y)
```

```python
interact(func,x=widgets.IntSlider(min=-10,max=30,step=1,value=0));
```

```python
interact(func,x=(-100,100,5))
```

```python
@interact(x=(0.0,20.0,0.5))
def h(x=5.0):
    return x
```

```python
interact(func,x=['hellow','op1','op2'])
```

```python
interact(func,x={'one':1,'two':2,'three':3})
```

```python
from IPython.display import display
def f(a,b):
    display(a+b)
    return a+b  
```

```python
w = interactive(f,a=10,b=20)
```
