LOGIC

```python
def is_even(num):
    return num % 2 == 0   
```

```python
is_even(20)
```

```python
is_even(21)
```

```python
#return true if any number is even in a list
def check_even_list(num_list):
    for num in num_list:
        if is_even(num):
            return True
    return False   
```

```python
check_even_list([1,3,5])
```

```python
check_even_list([1,1,1,112])
```
